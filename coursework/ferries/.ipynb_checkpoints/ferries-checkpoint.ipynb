```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

# Load the source data
Load all the port locations into a numpy array.
Also, set the number of boats, hillclimber iterations and hillclimber repeats.

```python
with open("stops.csv", "r") as srcF:
    ports = np.array(list(csv.reader(srcF, delimiter=",")), dtype=float)

boats = 3
iterations = 500
repeats = 30
```

# Generate a random solution
Start by generating a random solution.

```python
def genRandomSolution(ports, boats):
    order = np.random.permutation(range(len(ports)))
    return order.reshape((boats, int(len(order) / boats)))
base_sln = genRandomSolution(ports, boats)
print(base_sln)
```

# Create the evaluator and solution visualiser
The evaluator takes in the locations of the ports and the solution and calculates the spatial distance that the solution covers using `scipy.spatial.distance.euclidean`.

The solution visualiser takes in the locations of the ports and the solution and draws it onto a graph.

```python
from scipy.spatial.distance import euclidean

def evaluator(ports, sln):
    shape = sln.shape
    acc = 0

    for X in range(shape[0]):
        for Y in range(shape[1]):
            i1, i2 = Y, Y + 1
            if i2 == shape[1]:
                i2 = 0
            acc += euclidean(ports[sln[X][i1]], ports[sln[X][i2]])
    return acc

def graphRoutes(sln, ports, cMapName= 'viridis'):
    plt.figure()
    shape = sln.shape
    cmap = plt.cm.get_cmap(cMapName)

    for X in range(shape[0]):
        for Y in range(shape[1]):
            i1, i2 = Y, Y + 1
            if i2 == shape[1]:
                i2 = 0
            p1 = ports[sln[X][i1]]
            p2 = ports[sln[X][i2]]
            c = cmap(X / shape[0])
            plt.plot([p1[0], p2[0]], [p1[1], p2[1]], c=c)

    plt.scatter(x=ports[:, 0], y=ports[:, 1], c="y", zorder=10)

    plt.title("Ferry Routes")
    plt.xlabel("X grid reference")
    plt.ylabel("Y grid reference")

print("Base Solution fitness: ",evaluator(ports, base_sln))
print(graphRoutes(base_sln, ports))
```

# Create the three optimisers
These are the methods that will mutate the solution.

## Shuffle
This method takes in the solution, picks four contiguous points and randomly shuffles them.

```python
def shuffle(sln, group = 4):
    flat = np.ndarray.flatten(sln)
    index = np.random.randint(len(flat) - (group - 1))
    contigious = flat[index:index + group]
    shuffled = np.random.permutation(contigious)
    flat[index:index + group] = shuffled
    return flat.reshape(sln.shape)
print(shuffle(base_sln))
```

## Swap
This method takes in the solution, picks two different points and swaps them.

```python
def swap(sln):
    flat = np.ndarray.flatten(sln)
    flatLen = len(flat)
    index1 = np.random.randint(flatLen)
    index2 = np.random.randint(flatLen)
    while index1 == index2:
        index2 = np.random.randint(flatLen)
    flat[index1], flat[index2] = flat[index2], flat[index1]
    return flat.reshape(sln.shape)
print(swap(base_sln))
```

## Ruin and Recreate
This method takes in the solution and creates a novel solution in the same shape.

```python
def ruinAndRecreate(sln):
    length = len(np.ndarray.flatten(sln))
    perm = np.random.permutation(range(length))
    return np.reshape(perm, sln.shape)
print(ruinAndRecreate(base_sln))
```

# Create the Hill Climber and the Repeater
The hill climber is the method that will run the mutator and the evaluator and determine if the solution has improved (the distance it covers has decreased). If it has, it will be fed into the mutator in subsequent iterations. At the end, the current solution and its fitness will be returned.

The repeater will run the hill climber multiple times and collect the results. At the end, the fittest solution and the sitness data for all solutions will be returned.

```python
def hillClimber(sln, ports, mutator, max_iterations):
    currentFitness = evaluator(ports, sln)
    fitness = [currentFitness]
    for _ in range(max_iterations):
        newSln = mutator(sln)
        newFitness = evaluator(ports, newSln)
        if newFitness < currentFitness:
            sln = newSln
            currentFitness = newFitness
        fitness.append(currentFitness)
    return sln, fitness

def repeater(base_sln, ports, mutator, runs, max_iterations):
    slnAcc = []
    fitnessAcc = []
    for _ in range(runs):
        sln, fitness = hillClimber(base_sln, ports, mutator, max_iterations)
        slnAcc.append(sln)
        fitnessAcc.append(fitness)
    fitnessAcc = np.array(fitnessAcc)
    modelFitnessOrder = np.argsort(fitnessAcc[:, -1])
    return slnAcc[modelFitnessOrder[0]], fitnessAcc
```

```python
sln, shFitness = repeater(base_sln, ports, shuffle, repeats, iterations)
graphRoutes(sln, ports)
```

```python
sln, swFitness = repeater(base_sln, ports, swap, repeats, iterations)
graphRoutes(sln, ports)
```

```python
sln, ruFitness = repeater(base_sln, ports, ruinAndRecreate, repeats, iterations)
graphRoutes(sln, ports)
```

# Create the fitness visualiser
These graphs represent the minimum, mean and maximum distance each solution covers.

```python
def graphFitness(shFitnessAcc, swFitnessAcc, ruFitnessAcc, iterations, cMapName= 'viridis'):
    cmap = plt.cm.get_cmap(cMapName)
    shMaxAcc, shMinAcc, shMeanAcc = [], [], []
    swMaxAcc, swMinAcc, swMeanAcc = [], [], []
    ruMaxAcc, ruMinAcc, ruMeanAcc = [], [], []

    for X in range(iterations):
        shMaxAcc.append(max(shFitnessAcc[:, X]))
        shMinAcc.append(min(shFitnessAcc[:, X]))
        shMeanAcc.append(np.mean(shFitnessAcc[:, X]))

        swMaxAcc.append(max(swFitnessAcc[:, X]))
        swMinAcc.append(min(swFitnessAcc[:, X]))
        swMeanAcc.append(np.mean(swFitnessAcc[:, X]))

        ruMaxAcc.append(max(ruFitnessAcc[:, X]))
        ruMinAcc.append(min(ruFitnessAcc[:, X]))
        ruMeanAcc.append(np.mean(ruFitnessAcc[:, X]))

    plt.figure()
    plt.title("Maximum Distance over Iterations")
    plt.xlabel("Iterations")
    plt.ylabel("Distance (Lower is better)")
    plt.plot(shMaxAcc, c=cmap(0))
    plt.plot(swMaxAcc, c=cmap(0.5))
    plt.plot(ruMaxAcc, c=cmap(0.99))
    plt.legend(["Shuffle Maximum Distance", "Swap Maximum Distance", "Ruin & Repeat Maximum Distance"])

    plt.figure()
    plt.title("Mean Distance over Iterations")
    plt.xlabel("Iterations")
    plt.ylabel("Distance (Lower is better)")
    plt.plot(shMeanAcc, c=cmap(0))
    plt.plot(swMeanAcc, c=cmap(0.5))
    plt.plot(ruMeanAcc, c=cmap(0.99))
    plt.legend(["Shuffle Mean Distance", "Swap Mean Distance", "Ruin & Repeat Mean Distance"])

    plt.figure()
    plt.title("Minimum Distance over Iterations")
    plt.xlabel("Iterations")
    plt.ylabel("Distance (Lower is better)")
    plt.plot(shMinAcc, c=cmap(0))
    plt.plot(swMinAcc, c=cmap(0.5))
    plt.plot(ruMinAcc, c=cmap(0.99))
    plt.legend(["Shuffle Minimum Distance", "Swap Minimum Distance", "Ruin & Repeat Minimum Distance"])
```

```python
graphFitness(shFitness, swFitness, ruFitness, iterations)
```
